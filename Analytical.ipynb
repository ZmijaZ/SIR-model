```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
ndays = 100
h = 0.025
num_iters = int(ndays/h)
beta = 0.3
gamma = 1/14

S = np.zeros(num_iters)
I = np.zeros(num_iters)
R = np.zeros(num_iters)
t = np.arange(num_iters)*h
```

```python
I[0] = 0.001
S[0] = 1 - I[0]
R[0] = 0
```

```python
for i in range(num_iters-1):
    S[i+1] = S[i] - beta*(S[i]*I[i])*h
    I[i+1] = I[i] + (beta*S[i]*I[i]-gamma*I[i])*h
    R[i+1] = R[i] + (gamma*I[i])*h
```

```python
fig = plt.figure(10)
plt.plot(t, S)
plt.plot(t, I)
plt.plot(t, R)
```
